```python
import json

US_WOE_ID = 23424977

# Prefix ID with the underscore for query string parameterization.
# Without the underscore, the twitter package appends the ID value
# to the URL itself as a special case keyword argument.

#Current twitter trends
US_trends = twitter_api.trends.place(_id=US_WOE_ID)

#List of the top 10 trends from the current twitter trends data
US_top10 = US_trends[0]['trends'][:10]

print json.dumps(US_top10, indent=1)
```

```python
#Pulls trend name from top 10 list of trends
US_top10_set = set(trend['name'] for trend in US_top10)

#Turns set into list and strips leading 'u' from each trend name
US_top10_set=list(US_top10_set)
top10 = []                
for i in US_top10_set[:10]:
    A=  str(i).encode("utf-8")
    top10.append(A)
print top10

```

```python

q = top10

count = 50

# Retrieve top 50 tweets for each trend in the top 10 list
statuses=[]
for i in q : 
    
    search_results = twitter_api.search.tweets(q=i, count=count)
    statuses += search_results['statuses']
    print "Length of statuses", len(statuses)

print len(statuses)

```

```python
#Get words and filtered by Stop Words

status_texts = [status['text']
                for status in statuses]
    
words = [w
            for t in status_texts
                for w in t.split() ]

from nltk.corpus import stopwords
import string

stop_words = stopwords.words('english')
stop_words += ['.',',','"',"'",'!','?','rt','td',':',';','de',"it's",'en']

list_of_words =[word.lower() for word in words if word.lower() not in stop_words]

print list_of_words

```

```python
from collections import Counter

from prettytable import PrettyTable

pt = PrettyTable(field_names=["Words", 'Count']) #Set column names
c = Counter(list_of_words)
[ pt.add_row(kv) for kv in c.most_common()[:20] ]
pt.align["Words"], pt.align['Count'] = 'l', 'r' # Set column alignment
print pt
```
